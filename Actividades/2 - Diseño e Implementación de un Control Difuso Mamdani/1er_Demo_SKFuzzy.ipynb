```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt
```

```python
# Define UoD
x = np.arange(30, 80, 0.1)

# Definimos MFs usando Triangular MF
slow = fuzz.trimf(x, [30,30,50])
medium = fuzz.trimf(x, [30,60,70])
fast = fuzz.trimf(x, [60,80,80])

# Usamos Matplotlib para graficar
plt.figure()
plt.plot(x, slow, 'b', label='Slow speed')
plt.plot(x, medium, 'k', label='Medium speed')
plt.plot(x, fast, 'r', label='Fast speed')
plt.title('Speeds')
plt.ylabel('Membership')
plt.xlabel('KMs per hour')
plt.legend()
```

```python
# Definimos MFs usando Trapezoidal MF
slow = fuzz.trapmf(x, [20,30,40,50])
medium = fuzz.trapmf(x, [30,40,50,70])
fast = fuzz.trapmf(x, [60,70,80,80])

# Usamos Matplotlib para graficar
plt.figure()
plt.plot(x, slow, 'b', label='Slow speed')
plt.plot(x, medium, 'k', label='Medium speed')
plt.plot(x, fast, 'r', label='Fast speed')
plt.title('Speeds')
plt.ylabel('Membership')
plt.xlabel('KMs per hour')
plt.legend()
```

```python
# Definimos MFs usando Gaussian y GB
slow = fuzz.gaussmf(x, 40,5)
fast = fuzz.gbellmf(x, 8,4,50) # a=width, b=slope, c=center

# Usamos Matplotlib para graficar
plt.figure()
plt.plot(x, slow, 'b', label='Slow speed')
#plt.plot(x, medium, 'k', label='Medium speed')
plt.plot(x, fast, 'r', label='Fast speed')
plt.title('Speeds')
plt.ylabel('Membership')
plt.xlabel('KMs per hour')
plt.legend()
```
